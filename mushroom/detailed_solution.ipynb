```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('mushroom_train.csv')
test=pd.read_csv('mushroom_test.csv')
train.head()
```

```python
test.head()
```

```python
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for col in test.columns:
    if((col!='radius') and (col!='weight')):
        train[col]=encoder.fit_transform(train[col])
        test[col]=encoder.transform(test[col])

l_class=LabelEncoder()        
train['class']=l_class.fit_transform(train['class'])
            
```

```python
#for col in test.columns:
#    print(train.groupby(col).size())
```

```python
train=train.drop(['veil-type'],axis=1)
test=test.drop(['veil-type'],axis=1)
```

```python
for col in test.columns:
    if((col!='radius') and (col!='weight')):
        plt.figure(figsize=(5,5))
        arr=sns.violinplot(train[col],train['class'])
        print(train.groupby(col).size())
        plt.show()
    
```

```python
plt.figure(figsize=(5,5))
arr=sns.kdeplot(train['radius'],train['class'])
plt.show()
  
```

```python
train_final=train[['class','odor','spore-print-color','gill-size','cap-color','habitat']]
test_final=test[['odor','spore-print-color','gill-size','cap-color','habitat']]
```

What all I did above is what a tree classifier do.
So we can apply direct tree classification in our selected columns(or in whole dataset, both will give same result).

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_final.drop(['class'],axis=1),train_final['class'])
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score
extra=ExtraTreesClassifier(n_estimators=200,n_jobs=-1).fit(x_train,y_train).predict(x_test)
accuracy_score(y_test,extra)
```

```python
prediction=ExtraTreesClassifier(n_estimators=200,n_jobs=-1).fit(train_final.drop(['class'],axis=1),train_final['class']).predict(test_final)
```

```python

test['class']=l_class.inverse_transform(prediction)
```

```python
test.to_csv('prediction.csv',index=False)
```
